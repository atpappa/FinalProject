```python
import pandas as pd
```

```python
tc = pd.read_csv("TripleCrownRaces_2005-2019.csv")
tc
```

```python
toptc = tc[tc.final_place <= 5]
toptc
```

```python
count = toptc.groupby('Jockey').count().reset_index().rename(columns={'Jockey': 'Name', 'final_place': 'Count'})
countj = count[["Name", "Count"]]
countj
```

```python
countj= countj.sort_values(by=['Count'], ascending=False)
countj
```

```python
countj.to_csv("top5jockey.csv")
```

```python
countt = toptc.groupby('Trainer').count().reset_index().rename(columns={'Trainer': 'Name', 'final_place': 'Count'})
countt = countt[["Name", "Count"]]
countt
```

```python
countt= countt.sort_values(by=['Count'], ascending=False)
countt
```

```python
countt.to_csv("top5trainer.csv")
```
